```python
import matplotlib.pyplot as plt

%matplotlib inline

import numpy as np
import healpy as hp
```

```python
NSIDE = 32
print(
    "Approximate resolution at NSIDE {} is {:.2} deg".format(
        NSIDE, hp.nside2resol(NSIDE, arcmin=True) / 60
    )
)
```

```python
NPIX = hp.nside2npix(NSIDE)
print(NPIX)
```

```python
m = np.arange(NPIX)

theta1 = np.pi/2 - np.deg2rad(67.5) 
phi1 = np.deg2rad(264)

vec = hp.ang2vec(np.pi/2, np.pi*3 / 4)
print(vec)


hp.mollview(m, title="Mollview image RING",coord=['G'])
hp.graticule()
hp.projscatter(theta1, phi1,marker='*',s=30,color='k',coord=['G'])
hp.projscatter(np.pi/2, np.pi*3 / 4,marker='*',s=30,color='w',coord=['G'])
```

```python
def skyvec(theta,phi): 
    st = np.sin(theta)
    cp = np.cos(phi)
    sp = np.sin(phi)
    ct = np.cos(theta)
    return np.array([st*cp,st*sp,ct])

v = skyvec(theta1, phi1)

RAV, DecV = hp.vec2ang(v, lonlat=True)
RAV_Col, DecV_lon = hp.vec2ang(v)
print(RAV, DecV)
print(np.degrees(RAV_Col), np.degrees(DecV_lon))
```
