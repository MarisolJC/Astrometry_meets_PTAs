```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

#from matplotlib import rc
#rc('font', **{'family':'serif','serif':'Latin Modern Roman'})
#rc('text', usetex=True)

font = {'size': 13.5, 'family':'Serif'}
#font = {'size': 18, 'family':'Latin Modern Roman'}
axislabelfontsize='large'
matplotlib.rc('font', **font)
matplotlib.rc('text', usetex=True)

cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
def Trace_H0H0(zeta):
    Trace_H0H0 = (np.pi**2 * (1 + np.cos(zeta)**2) * 
              (5 - 2 * np.cos(zeta) - 7 * np.cos(zeta)**2 + 
              6 * (1 - np.cos(zeta))**2 * np.log(np.sin(zeta / 2)**2))**2) / \
             (9 * (1 + np.cos(zeta))**2)
    return Trace_H0H0

def Trace_H1H1_negV(zeta):
    Trace_H1H1 = (4 * np.pi**2 * (-1 + np.cos(zeta))**2 * (1 + np.cos(zeta)**2) * 
          (-2 - np.cos(zeta) + np.cos(zeta)**2 + np.log(8) - 
           np.cos(zeta) * np.log(8) + 
           3 * (-1 + np.cos(zeta)) * np.log(1 - np.cos(zeta)))**2) / \
         (9 * (1 + np.cos(zeta))**2)
    
    return Trace_H1H1

def Trace_H1H1_V(zeta):
    Trace_H1H1 = (1 / 18) * np.pi**2 * (3 + np.cos(2 * zeta)) * \
            (-3 + np.cos(2 * zeta) + np.log(64) - 
             6 * np.log(1 - np.cos(zeta)) + 
             np.cos(zeta) * (-2 + 6 * np.log(np.sin(zeta / 2)**2)))**2 * \
            np.tan(zeta / 2)**4
    return Trace_H1H1

def K0K0(zeta):
    K0K0 = -((16 * np.pi**2 * (-1 + np.cos(zeta)) * 
            (-2 - 2 * np.cos(zeta) + np.log(8) + 
             3 * np.cos(zeta) * np.log(0.5 * (1 - np.cos(zeta))) - 
             3 * np.log(1 - np.cos(zeta)))**2) / 
           (9 * (1 + np.cos(zeta))))
    return K0K0

def K1K1(zeta):
    K1K1 = (1/9)*np.pi**2 * (5 + 2*np.cos(zeta) - 3*np.cos(2*zeta) - \
            12*(-1 + np.cos(zeta))*np.log( np.sin(zeta/2)**2 ) )**2 * (np.tan(zeta/2)**2)
    return K1K1
    
```

```python
zeta = np.arange(0.0001, np.pi, 0.01)
zeta_deg = np.rad2deg(zeta)

#min = np.min(zeta_deg)
#max = np.max(zeta_deg)

plt.figure(figsize=(6, 4))
plt.plot(zeta_deg, Trace_H0H0(zeta), color='r')
#plt.title('$Tr[H_{0} H_{0}]$')
plt.xlabel(r'$\mathrm{Angle} (\zeta)$',size=18)
plt.ylabel(r'$\mathrm{Tr}[\mathbf{H}_0 \mathbf{H}_0]$',size=18)
#plt.grid()

plt.savefig("Trace_MonMon.pdf", bbox_inches='tight')
```
