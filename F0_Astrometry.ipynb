```python
import numpy as np
import matplotlib
from matplotlib import colormaps
import matplotlib.pyplot as plt

from matplotlib import rc
rc('font', **{'family':'serif','serif':'Latin Modern Roman'})
rc('text', usetex=True)
```

```python
def hola(y):
    return np.where(y == 0, 0, np.log(y))
    
def fF0(zeta):
    cos_zeta = np.cos(zeta)
    y = (1- cos_zeta)/2
    num = np.pi*(1- 8*y +7*(y**2) - 6*(y**2)*hola(y))
    den = 3* ((1-y)**2)
    return num/den #num/den

def fF2(zeta):
    cos_zeta = np.cos(zeta)
    y = (1- cos_zeta)/2
    num = np.pi*y*((y-1)*(2*y +1) - 3*y*hola(y))
    den = 3*((y-1)**2)
    return num/den #num/den
```

```python
zeta_deg = np.linspace(0.000001, 179.9, 500)
zeta_rad = np.deg2rad(zeta_deg)

fF0(zeta_rad)
fF2(zeta_rad)

plt. plot(zeta_deg, abs(fF0(zeta_rad)), label='$|F_0|$')
plt. plot(zeta_deg, abs(fF2(zeta_rad)), label='$|F_2|$')
plt.legend()
plt.title('$|F_{0}|, |F_{2}|$')
plt.xlabel(r'Angle $(\xi)$')
plt.ylabel(r'')
plt.axhline(y=0, color='red', linestyle='--', label='y = 0.5')
plt.axvline(x=95, color='blue', linestyle='--', label='x = 95Â°')
plt.savefig('abs_F0_F2.pdf')
```

```python
plt. plot(zeta_deg, fF0(zeta_rad), label='$F_0$')
plt. plot(zeta_deg, fF2(zeta_rad), label='$F_2$')
plt.legend()
plt.title('$F_{0}, F_{2}$')
plt.xlabel(r'Angle $(\xi)$')
plt.ylabel(r'')
plt.axhline(y=0, color='red', linestyle='--', label='y = 0.5')
plt.axvline(x=95, color='blue', linestyle='--', label='x = 95Â°')
plt.savefig('F0_F2.pdf')
```
